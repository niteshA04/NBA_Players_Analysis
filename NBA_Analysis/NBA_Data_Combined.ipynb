# Import Libraries

```python
import pandas as pd
```

# Load Data

```python
d1 = pd.read_csv('2012_nba_draft_combine.csv', index_col=0)
d2 = pd.read_csv('2013_nba_draft_combine.csv', index_col=0)
d3 = pd.read_csv('2014_nba_draft_combine.csv', index_col=0)
d4 = pd.read_csv('2015_nba_draft_combine.csv', index_col=0)
d5 = pd.read_csv('2016_nba_draft_combine.csv', index_col=0)
```

# Combining Data

```python
data = pd.concat([d1, d2, d3, d4, d5], ignore_index=True)

data.head(20)
```

```python
data.isna().sum()
```

```python
data['Draft pick'].fillna(data['Draft pick'].median(), inplace=True)
```

```python
data.isna().sum()
```

```python
data['Vertical (No Step)'].fillna(data['Vertical (Max)'].median(), inplace=True)
```

```python
data['Vertical (No Step Reach)'].fillna(data['Vertical (Max)'].median(), inplace=True)
```

```python
data['Weight'].fillna(data['Vertical (Max)'].median(), inplace=True)
```

```python
data['Body Fat'].fillna(data['Vertical (Max)'].median(), inplace=True)
```

```python
data['Bench'].fillna(data['Vertical (Max)'].median(), inplace=True)
```

```python
data['Agility'].fillna(data['Vertical (Max)'].median(), inplace=True)
data['Sprint'].fillna(data['Vertical (Max)'].median(), inplace=True)
```
